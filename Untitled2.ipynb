```python
y=df['Class']
```

```python
x=df.drop(columns=['Class'])
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3)
```

```python
from sklearn.preprocessing import MinMaxScaler
m1 = MinMaxScaler()
m2 = MinMaxScaler()
x_train = m1.fit_transform(x_train)
x_test = m1.transform(x_test)
y_train = m2.fit_transform(y_train.values.reshape(-1,1))
y_test = m2.transform(y_test.values.reshape(-1,1))
```

```python
x_train
```

```python
y_train
```

```python
from sklearn.tree import DecisionTreeClassifier
clf_gini=DecisionTreeClassifier(criterion="gini",max_depth=4,random_state=0)
clf_gini.fit(x_train,y_train)
```

```python
y_pred_gini=clf_gini.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred_gini)
```

```python
y_pred_train=clf_gini.predict(x_train)
```

```python
accuracy_score(y_pred_train,y_train)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(12,6))
from sklearn import tree
tree.plot_tree(clf_gini.fit(x_train,y_train))
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3)
```

```python
model = DecisionTreeClassifier()
model.get_params()
```

```python
param_grid ={
    'criterion':['gini','entropy'],
    'max_depth':[3,4,7]
}
```

```python
from sklearn.model_selection import GridSearchCV
grid_search = GridSearchCV(estimator=model,param_grid=param_grid,cv=4)
```

```python
grid_search.fit(x_train,y_train)
```

```python
best_params=grid_search.best_params_
best_model=grid_search.best_estimator_
```

```python
y_pred = best_model.predict(x_test)
accuracy = accuracy_score(y_test, y_pred)
print("Best Parameters:", best_params)
print("Test Accuracy:", accuracy)
```
